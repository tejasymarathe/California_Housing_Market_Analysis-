**Question [5]**
Create a data set by filling the missing data with the mean value of the total_bedrooms in the original data set.

```python
import pandas as pd    #Imported pandas library
df = pd.read_excel("housing.xlsx")    #Reading the uploaded database
print("Null values in 'total_bedrooms' column of the original dataset\n",(df["total_bedrooms"].isnull().sum()))     #Checking the null values before filling the null values with mean value
mean_value = df["total_bedrooms"].mean()    #Calculating mean value for total_bedrooms column
print("\nMean value of 'total_bedrooms' column is ", mean_value)    #Printing mean
df["total_bedrooms"] = df["total_bedrooms"].fillna(mean_value)     #Filling the null values with mean value in the total_bedrooms column
df.to_excel('no_null.xlsx', index=False, header=True)      #Exporting the updated data into new excel file
no_null = pd.read_excel("no_null.xlsx")     #Reading newly created excel file
print("\nNull values in 'total_bedrooms' column of the newly created dataset\n",(no_null["total_bedrooms"].isnull().sum()))     ##Checking the null values after filling the null values with mean value
```

**Question [6]**
Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
import pandas as pd     #Imported pandas library

class HousingData:      #Creating the class
    def __init__(self, excel_file):     #Defining the init method
        self.dataset = pd.read_excel("housing.xlsx")     #Reading the excel file

    def median(self, column_name):     #Defining the method
        median = self.dataset[column_name].median()      #calculating the median and storing it
        return f"The median of '{column_name}' column is {median}"     #Returning the median value with string

column = input("Enter the name of the column to find the median value: ")     #Taking user input for column name
median_calculator = HousingData(pd.read_excel("housing.xlsx"))     #Creating an instance which reads excel file
print(median_calculator.median(column))    #printing the median value by using instance
```

**Question [10]**
Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small.
If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
import pandas as pd    #Imported pandas library
df = pd.read_excel("housing.xlsx")    #Reading the uploaded database
df['total_bedroom_size'] = ''    #created empty column

#Applying conditions using loc
df.loc[df['total_bedrooms'] <= 10, 'total_bedroom_size'] = 'small'
df.loc[(df['total_bedrooms'] >= 11) & (df['total_bedrooms'] < 1000), 'total_bedroom_size'] = 'medium'
df.loc[df['total_bedrooms'] >= 1000, 'total_bedroom_size'] = 'large'

df.head() #Getting the filtered data
```
